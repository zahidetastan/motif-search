```python
import random  
import string  
def specific_string(length):  
    sample_string = 'agct' # define the specific string  
    # define the condition for random string  
    #random.seed(26)
    result = ''.join((random.choice(sample_string)) for x in range(length))  
    return result
kmer = specific_string(10)
kmer = kmer.upper()
kmer
```

```python
def mutation(kmerr):
    genom = specific_string(1)
    rnd = random.randint(0,val-1)
    
    mer = list(kmerr)
    
    if(mer[rnd].upper()==genom.upper()):
        return mutation(kmerr)
    
    if(mer[rnd].isupper()):
        mer[rnd] = genom
        kmerr = "".join(mer)
        return kmerr
    else:
        return mutation(kmerr)
```

```python
def addKmerToString(line):
    n = random.randint(0,490)
    n2 = n+10
    newmer = kmer
    for i in range(4):
        newmer = mutation(newmer)
    line = line.replace(line[n:n2],newmer)
    return line
```

```python
def ScoreMotifs(Motifs):
    summation = 0
    Scores = []
    Count = []
    Length = len(Motifs)
    for Col in range(len(Motifs[0])):
        Count.append({'A':0,'C':0,'G':0,'T':0}) 
        for Row in range(Length):
            Count[Col][Motifs[Row][Col].upper()] += 1
    for Score in Count:
        PredominantLetter = max(Score, key=lambda i: Score[i])
        Total = sum(Score.values())
        Scores.append(Total - Score[PredominantLetter])
        
    for i in range(0, len(Scores)):    
        summation = summation + Scores[i]; 
    return summation
```

```python
def CountMotifs(Motifs):
    Counts = []
    for Col in range(len(Motifs[0])):
        Count = {'A':0,'T':0,'C':0,'G':0}
        for Motif in Motifs:
            Count[Motif[Col].upper()] += 1
        Counts.append(Count)
    return Counts
```

```python
def ProfileMotifs(Motifs):
    Profiles = []
    Counts = CountMotifs(Motifs)
    for Count in Counts:
        Profile = {}
        Total = sum(Count.values())
        for Letter in bases:
            Profile[Letter] = Count[Letter] / Total
        Profiles.append(Profile)
    return Profiles
```

```python
def ConsensusMotif(Motifs):
    Motif = []
    Profiles = ProfileMotifs(Motifs)
    for Profile in Profiles:
        PredominantLetter = max(Profile, key=lambda i: Profile[i])
        Motif.append(PredominantLetter)
    return Motif
```

```python
import random

def PlusOneCountMotifs(Motifs):
    counts = []
    for column in range(len(Motifs[0]) ):
        count = {'A':1,'T':1,'C':1,'G':1}
        for Motif in Motifs:
            count[Motif[column].upper()] += 1
        counts.append(count)
    return counts

bases = ['A', 'C', 'G', 'T']

def UpdatedProfileMotifs(Motifs):
    profiles = []
    countsArr = PlusOneCountMotifs(Motifs)
    for count in countsArr:
        profile = {}
        summation = sum(count.values())
        for base in bases:
            profile[base] = int((count[base] / summation) * 100) + random.randint(0,1)
        profiles.append(profile)
    return profiles
```

```python
def MotifsFromProfile(profile, lines):
    motifs = []
    k = len(profile)
    for line in lines:
        kmer = MostProbableKmer(line, k, profile)
        motifs.append(kmer)
    return motifs
```

```python
import numpy as np

def GibbsSampler(lines, N):
    motifs = []
    motifs = RandomMotifs() #Motifs array takes the random motifs on each line. 
    
    bestMotifs = motifs #initiliaze bestmotifs as motifs.
    for val in range(N):
        i = np.random.randint(0, len(lines)-1)
        profile = UpdatedProfileMotifs(motifs) #Take the updated Profile Matrix as Profile
        motifs[i] = MotifsFromProfile(profile, lines)[i]
        if ScoreMotifs(motifs) < ScoreMotifs(bestMotifs):
            bestMotifs = motifs
    return bestMotifs
```

```python
def MostProbableKmer(line, k, Profile):
    kmers = []
    mostProbable = line[0:k]
    for i in range((len(line)+1) - k):
        kmers.append(line[i:i+k])
    probability = 0
    for kmer in kmers:
        newProb = ProbabilityMotif(kmer, Profile)
        if newProb > probability:
            mostProbable = kmer
            probability = newProb
    return mostProbable
```

```python
def ProbabilityMotif(Pattern, Profile):
    Probabilities = []
    for Index, Letter in enumerate(Pattern):
        Probability = Profile[Index][Letter]
        Probabilities.append(Probability)
    return reduce(multiply, Probabilities, 1)
```

```python
def multiply(x,y):
    return x*y
```

```python
def RandomizedMotifSearch(lines):  
    motifs = []
    motifs = RandomMotifs() #Motifs array takes the random motifs on each line. 
    
    bestMotifs = motifs #initiliaze bestmotifs as motifs.
    while True:
        profile = UpdatedProfileMotifs(motifs)
        motifs = MotifsFromProfile(profile, lines)
        if ScoreMotifs(motifs) < ScoreMotifs(bestMotifs):
            bestMotifs = motifs
        else:
            return bestMotifs
```
