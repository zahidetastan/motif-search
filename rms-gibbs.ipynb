```python
import random  
import string  
def specific_string(length):  
    sample_string = 'agct' # define the specific string  
    # define the condition for random string  
    #random.seed(26)
    result = ''.join((random.choice(sample_string)) for x in range(length))  
    return result
kmer = specific_string(10)
kmer = kmer.upper()
kmer
```

```python
def mutation(kmerr):
    genom = specific_string(1)
    rnd = random.randint(0,9)
    
    mer = list(kmerr)
    
    if(mer[rnd].upper()==genom.upper()):
        return mutation(kmerr)
    
    if(mer[rnd].isupper()):
        mer[rnd] = genom
        kmerr = "".join(mer)
        return kmerr
    else:
        return mutation(kmerr)
```

```python
def addKmerToString(line,rangeArr):
    n = random.randint(0,490)
    n2 = n+10
    if(n not in rangeArr):
        newmer = kmer
        for i in range(4):
            newmer = mutation(newmer)
        line = line.replace(line[n:n2],newmer)
        return line
    else:
        return addKmerToString(line, rangeArr)
```

```python
cnt = 0
for line in lines:
    line = line.rstrip("\n")        

    rangeArr = []
    for i in range(4):
        line = addKmerToString(line,rangeArr)
    lines[cnt] = line
    print(line)
    cnt += 1
```
